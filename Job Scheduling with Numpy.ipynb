```python
import numpy as np
import pandas as pd

# read excel file
jobs = pd.read_excel(r"C:\Users\selin\Desktop\jobs.xlsx", index_col = 0)
jobs
```

```python
# Create new columns for functions
jobs["Flow Time"] = np.nan
jobs["Tardiness"] = np.nan
jobs
```

```python
# First Come First Serve Function, creating a loop

start_time = 0
for index, row in jobs.iterrows():
    jobs.loc[index, "Flow Time"] = start_time + jobs.loc[index, "Processing Time"]
    start_time = jobs.loc[index, "Flow Time"]
    jobs.loc[index, "Tardiness"] = max(0, jobs.loc[index, "Flow Time"] - jobs.loc[index, "Due Date"])
print(jobs)
```

```python
# defining a function
def fcfs(jobs):
    start_time = 0
    for index, row in jobs.iterrows():
        jobs.loc[index, "Flow Time"] = start_time + jobs.loc[index, "Processing Time"]
        start_time = jobs.loc[index, "Flow Time"]
        jobs.loc[index, "Tardiness"] = max(0, jobs.loc[index, "Flow Time"] - jobs.loc[index, "Due Date"])
    return jobs

fcfs(jobs)
```

```python
# Alternative function for 
def fcfs_alt(jobs):
    for index, row in jobs.iterrows():
        jobs["Flow Time"] = jobs["Processing Time"].cumsum()
        jobs["Tardiness"] = np.maximum(0,jobs["Flow Time"]-jobs["Due Date"])
    return jobs

fcfs_alt(jobs)
```

```python
# Defining performance criteria
def performance(jobs):
    result ={"Average job flow time": jobs["Flow Time"].mean(),
             "Average Tardiness": jobs["Tardiness"].mean(),
             "Make span": jobs["Flow Time"].max(),
             "Average Nummber of Jobs": jobs["Flow Time"].sum() / jobs["Flow Time"].max()}
    return pd.Series(result)

print(performance(jobs))
```

```python
# Sort the dataframe for shortest procesing time scheduling
sorted_jobs = jobs.sort_values("Processing Time")
sorted_jobs
```

```python
def spt(sorted_jobs):
    start_time = 0
    for index, row in sorted_jobs.iterrows():
        sorted_jobs.loc[index, "Flow Time"] = start_time + sorted_jobs.loc[index, "Processing Time"]
        start_time = sorted_jobs.loc[index, "Flow Time"]
        sorted_jobs.loc[index, "Tardiness"] = max(0, sorted_jobs.loc[index, "Flow Time"] - sorted_jobs.loc[index, "Due Date"])
    return sorted_jobs

spt(sorted_jobs)
```
