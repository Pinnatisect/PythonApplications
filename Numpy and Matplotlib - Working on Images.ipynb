```python
# image as grayscale
f = misc.face(gray = True)
print(f.shape)
plt.imshow(f, cmap = "gray")
```

```python
import numpy as np
k = np.zeros((2,6))
print(k)
```

```python
k = 20
resized_image = np.zeros((f.shape[0]//k,f.shape[1]//k))

for i in range(resized_image.shape[0]):
    for j in range(resized_image.shape[1]):
        resized_image[i,j] = int(np.mean(f[k*i:k*(i+1), k*j:k*(j+1)]))
print(resized_image.shape)
plt.imshow(resized_image,cmap = "gray")
```

```python
blur_image = resized_image.copy()
k = 4
for i in range(resized_image.shape[0]-k+1):
    for j in range(resized_image.shape[1]-k+1):
        blur_image[i:(i+k), j: (j+k)] = int(np.mean(resized_image[i:(i+k), j: (j+k)]))
plt.imshow(blur_image, cmap = "gray")
```
